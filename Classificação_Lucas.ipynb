```python
import numpy as np
from sklearn.datasets import load_iris

# Fun√ß√£o para calcular a dist√¢ncia euclidiana
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# Classe do classificador KNN
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        # Computa as dist√¢ncias
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]

        # Obt√©m os k vizinhos mais pr√≥ximos
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # Obt√©m o r√≥tulo de classe mais comum
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        return most_common

# Carregar a base de dados Iris
print(load_iris())
iris = load_iris()
X = iris.data
y = iris.target
class_names = iris.target_names  # Nomes das classes
```

```python
class Perceptron:
    # O m√©todo __init__ inicializa o Perceptron com uma taxa de aprendizado e um n√∫mero de itera√ß√µes.
    def __init__(self, learning_rate=0.01, n_iter=1000):
        # Armazena a taxa de aprendizado (learning_rate) fornecida.
        self.learning_rate = learning_rate
        # Armazena o n√∫mero de itera√ß√µes (n_iter) que o algoritmo executar√°.
        self.n_iter = n_iter

    # O m√©todo fit treina o modelo Perceptron nos dados de entrada X e nos r√≥tulos correspondentes y.
    def fit(self, X, y):
        # Armazena o n√∫mero de amostras (n_samples) e o n√∫mero de caracter√≠sticas (n_features) dos dados de entrada X.
        self.n_samples, self.n_features = X.shape
        # Inicializa os pesos (weights) como um array de zeros, com o mesmo n√∫mero de elementos que as caracter√≠sticas.
        self.weights = np.zeros(self.n_features)
        # Inicializa o bias (termo de polariza√ß√£o) como zero.
        self.bias = 0
        # Inicializa uma lista vazia para armazenar o n√∫mero de erros em cada itera√ß√£o.
        self.errors_ = []

        # Loop principal do algoritmo de aprendizado, que ser√° executado pelo n√∫mero de itera√ß√µes especificado.
        for _ in range(self.n_iter):
            # Inicializa o contador de erros para a itera√ß√£o atual.
            errors = 0
            # Itera sobre cada amostra x_i e seu respectivo r√≥tulo y[idx].
            for idx, x_i in enumerate(X):
                # Calcula a sa√≠da linear, que √© o produto escalar dos pesos e a amostra, somado ao bias.
                linear_output = np.dot(x_i, self.weights) + self.bias
                # Aplica a fun√ß√£o de ativa√ß√£o (step function) √† sa√≠da linear para obter a previs√£o (predicted).
                predicted = self._step_function(linear_output)
                # Calcula o valor de atualiza√ß√£o, que depende da diferen√ßa entre o r√≥tulo real e a previs√£o.
                update = self.learning_rate * (y[idx] - predicted)
                # Atualiza os pesos multiplicando o valor de atualiza√ß√£o pela amostra.
                self.weights += update * x_i
                # Atualiza o bias somando o valor de atualiza√ß√£o.
                self.bias += update
                # Incrementa o contador de erros se houver uma atualiza√ß√£o (update n√£o for zero).
                errors += int(update != 0.0)
            # Adiciona o n√∫mero de erros da itera√ß√£o atual √† lista de erros.
            self.errors_.append(errors)

    # A fun√ß√£o step_function aplica a fun√ß√£o de ativa√ß√£o degrau (step function) √† entrada x.
    def _step_function(self, x):
        # Retorna 1 se x for maior ou igual a 0, caso contr√°rio, retorna 0.
        return np.where(x >= 0, 1, 0)

    # O m√©todo predict faz previs√µes para um conjunto de dados X, usando o modelo treinado.
    def predict(self, X):
        # Calcula a sa√≠da linear para cada amostra em X.
        linear_output = np.dot(X, self.weights) + self.bias
        # Aplica a fun√ß√£o de ativa√ß√£o e retorna as previs√µes.
        return self._step_function(linear_output)

# Cria uma inst√¢ncia do Perceptron com uma taxa de aprendizado de 0.01 e 1000 itera√ß√µes.
perceptron = Perceptron(learning_rate=0.01, n_iter=1000)

# Treina o Perceptron nos dados de entrada X_binary e nos r√≥tulos correspondentes y_binary.
perceptron.fit(X_binary, y_binary)

# Faz previs√µes usando os dados de entrada X_binary.
predictions = perceptron.predict(X_binary)

# Calcula a acur√°cia das previs√µes comparando-as com os r√≥tulos reais y_binary.
accuracy = np.mean(predictions == y_binary)

# Imprime a acur√°cia.
print("Acur√°cia:", accuracy)
```

```python
from sklearn.preprocessing import StandardScaler

# Normalizar os dados
scaler = StandardScaler()
# Ajusta o normalizador aos dados X e transforma X para ter m√©dia zero e vari√¢ncia unit√°ria.
X = scaler.fit_transform(X)

# Usar apenas as duas primeiras classes para a classifica√ß√£o bin√°ria
# Cria uma m√°scara booleana para selecionar apenas os √≠ndices onde y √© menor que 2 (classes 0 e 1).
binary_indices = y < 2
# Filtra X para incluir apenas as amostras que pertencem √†s duas primeiras classes.
X_binary = X[binary_indices]
# Filtra y para incluir apenas os r√≥tulos que pertencem √†s duas primeiras classes.
y_binary = y[binary_indices]

# Implementa√ß√£o do Perceptron com a Regra Delta
class PerceptronDelta:
    # O m√©todo __init__ inicializa o PerceptronDelta com uma taxa de aprendizado e um n√∫mero de itera√ß√µes.
    def __init__(self, learning_rate=0.01, n_iter=1000):
        # Armazena a taxa de aprendizado (learning_rate) fornecida.
        self.learning_rate = learning_rate
        # Armazena o n√∫mero de itera√ß√µes (n_iter) que o algoritmo executar√°.
        self.n_iter = n_iter

    # O m√©todo fit treina o modelo PerceptronDelta nos dados de entrada X e nos r√≥tulos correspondentes y.
    def fit(self, X, y):
        # Armazena o n√∫mero de amostras (n_samples) e o n√∫mero de caracter√≠sticas (n_features) dos dados de entrada X.
        self.n_samples, self.n_features = X.shape
        # Inicializa os pesos (weights) como um array de zeros, com o mesmo n√∫mero de elementos que as caracter√≠sticas.
        self.weights = np.zeros(self.n_features)
        # Inicializa o bias (termo de polariza√ß√£o) como zero.
        self.bias = 0
        # Inicializa uma lista vazia para armazenar o n√∫mero de erros em cada itera√ß√£o.
        self.errors_ = []

        # Loop principal do algoritmo de aprendizado, que ser√° executado pelo n√∫mero de itera√ß√µes especificado.
        for _ in range(self.n_iter):
            # Inicializa o erro total para a itera√ß√£o atual.
            total_error = 0
            # Itera sobre cada amostra x_i e seu respectivo r√≥tulo y[idx].
            for idx, x_i in enumerate(X):
                # Calcula a sa√≠da linear, que √© o produto escalar dos pesos e a amostra, somado ao bias.
                linear_output = np.dot(x_i, self.weights) + self.bias
                # Aplica a fun√ß√£o de ativa√ß√£o (step function) √† sa√≠da linear para obter a previs√£o (predicted).
                predicted = self._step_function(linear_output)
                # Calcula o erro como a diferen√ßa entre o r√≥tulo real e a previs√£o.
                error = y[idx] - predicted
                # Atualiza os pesos multiplicando o erro pela amostra e pela taxa de aprendizado.
                self.weights += self.learning_rate * error * x_i
                # Atualiza o bias somando o erro multiplicado pela taxa de aprendizado.
                self.bias += self.learning_rate * error
                # Incrementa o erro total se houver erro na previs√£o.
                total_error += int(error != 0)
            # Adiciona o erro total da itera√ß√£o atual √† lista de erros.
            self.errors_.append(total_error)

    # A fun√ß√£o step_function aplica a fun√ß√£o de ativa√ß√£o degrau (step function) √† entrada x.
    def _step_function(self, x):
        # Retorna 1 se x for maior ou igual a 0, caso contr√°rio, retorna 0.
        return np.where(x >= 0, 1, 0)

    # O m√©todo predict faz previs√µes para um conjunto de dados X, usando o modelo treinado.
    def predict(self, X):
        # Calcula a sa√≠da linear para cada amostra em X.
        linear_output = np.dot(X, self.weights) + self.bias
        # Aplica a fun√ß√£o de ativa√ß√£o e retorna as previs√µes.
        return self._step_function(linear_output)

# Cria uma inst√¢ncia do PerceptronDelta com uma taxa de aprendizado de 0.01 e 1000 itera√ß√µes.
perceptronDelta = PerceptronDelta(learning_rate=0.01, n_iter=1000)

# Treina o PerceptronDelta nos dados de entrada X_binary e nos r√≥tulos correspondentes y_binary.
perceptronDelta.fit(X_binary, y_binary)

# Faz previs√µes usando os dados de entrada X_binary.
predictions = perceptronDelta.predict(X_binary)

# Calcula a acur√°cia das previs√µes comparando-as com os r√≥tulos reais y_binary.
accuracy = np.mean(predictions == y_binary)

# Imprime a acur√°cia.
print("Acur√°cia:", accuracy)
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Fun√ß√£o para plotar a superf√≠cie de decis√£o
def plot_decision_regions(X, y, classifier, title, resolution=0.02):
    # Configura√ß√£o do marcador e mapa de cores
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # Superf√≠cie de decis√£o
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Plot das amostras
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=cl, edgecolor='black')
    plt.title(title)
    plt.xlabel('Caracter√≠stica 1')
    plt.ylabel('Caracter√≠stica 2')
    plt.legend(loc='upper left')
    plt.show()

# Implementa√ß√£o do Perceptron
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.learning_rate = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):
        self.n_samples, self.n_features = X.shape
        self.weights = np.zeros(self.n_features)
        self.bias = 0
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                predicted = self._step_function(linear_output)
                update = self.learning_rate * (y[idx] - predicted)
                self.weights += update * x_i
                self.bias += update
                errors += int(update != 0.0)
            self.errors_.append(errors)

    def _step_function(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self._step_function(linear_output)

# Implementa√ß√£o do Perceptron com Regra Delta
class PerceptronDelta:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.learning_rate = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):
        self.n_samples, self.n_features = X.shape
        self.weights = np.zeros(self.n_features)
        self.bias = 0
        self.errors_ = []

        for _ in range(self.n_iter):
            total_error = 0
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                predicted = self._step_function(linear_output)
                error = y[idx] - predicted
                self.weights += self.learning_rate * error * x_i
                self.bias += self.learning_rate * error
                total_error += int(error != 0)
            self.errors_.append(total_error)

    def _step_function(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self._step_function(linear_output)

# Implementa√ß√£o do KNN
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        return most_common

# Fun√ß√£o para calcular a dist√¢ncia euclidiana
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# Carregar a base de dados Iris
iris = load_iris()
X = iris.data
y = iris.target

# Normalizar os dados
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Casos separ√°veis: Iris Setosa vs Versicolor (classes 0 e 1)
binary_indices_sv = y < 2
X_setosa_versicolor = X[binary_indices_sv]
y_setosa_versicolor = y[binary_indices_sv]

# Treinar e avaliar o Perceptron para Setosa vs Versicolor
perceptron_2d = Perceptron(learning_rate=0.01, n_iter=1000)
perceptron_2d.fit(X_setosa_versicolor[:, :2], y_setosa_versicolor)
predictions_sv_2d = perceptron_2d.predict(X_setosa_versicolor[:, :2])
accuracy_sv_2d = np.mean(predictions_sv_2d == y_setosa_versicolor)
print("Acur√°cia Perceptron (Setosa vs Versicolor) com 2 caracter√≠sticas:", accuracy_sv_2d)

# Treinar e avaliar o KNN para Setosa vs Versicolor
knn_sv = KNN(k=3)
knn_sv.fit(X_setosa_versicolor[:, :2], y_setosa_versicolor)
predictions_knn_sv = knn_sv.predict(X_setosa_versicolor[:, :2])
accuracy_knn_sv = np.mean(predictions_knn_sv == y_setosa_versicolor)
print("Acur√°cia KNN (Setosa vs Versicolor):", accuracy_knn_sv)

# Plotar a superf√≠cie de decis√£o para o Perceptron
plot_decision_regions(X_setosa_versicolor[:, :2], y_setosa_versicolor, perceptron_2d, "Superf√≠cie de Decis√£o Perceptron (Setosa vs Versicolor)")

# Casos n√£o separ√°veis: Iris Setosa vs Virginica (classes 0 e 2)
binary_indices_sv2 = (y == 0) | (y == 2)
X_setosa_virginica = X[binary_indices_sv2]
y_setosa_virginica = y[binary_indices_sv2]

# Treinar e avaliar o Perceptron com Regra Delta para Setosa vs Virginica
perceptron_delta = PerceptronDelta(learning_rate=0.01, n_iter=1000)
perceptron_delta.fit(X_setosa_virginica[:, :2], y_setosa_virginica)
predictions_sv2_delta = perceptron_delta.predict(X_setosa_virginica[:, :2])
accuracy_sv2_delta = np.mean(predictions_sv2_delta == y_setosa_virginica)
print("Acur√°cia Perceptron com Regra Delta (Setosa vs Virginica):", accuracy_sv2_delta)

# Treinar e avaliar o KNN para Setosa vs Virginica
knn_sv2 = KNN(k=3)
knn_sv2.fit(X_setosa_virginica[:, :2], y_setosa_virginica)
predictions_knn_sv2 = knn_sv2.predict(X_setosa_virginica[:, :2])
accuracy_knn_sv2 = np.mean(predictions_knn_sv2 == y_setosa_virginica)
print("Acur√°cia KNN (Setosa vs Virginica):", accuracy_knn_sv2)
```

```python
# Fun√ß√£o para calcular a matriz de confus√£o
def confusion_matrix_manual(y_true, y_pred):
    TP = TN = FP = FN = 0
    for true, pred in zip(y_true, y_pred):
        if true == 1 and pred == 1:
            TP += 1
        elif true == 0 and pred == 0:
            TN += 1
        elif true == 0 and pred == 1:
            FP += 1
        elif true == 1 and pred == 0:
            FN += 1
    return TP, TN, FP, FN

# Fun√ß√£o para calcular precis√£o
def precision_manual(TP, FP):
    return TP / (TP + FP) if (TP + FP) != 0 else 0

# Fun√ß√£o para calcular revoca√ß√£o (sensibilidade)
def recall_manual(TP, FN):
    return TP / (TP + FN) if (TP + FN) != 0 else 0

# Fun√ß√£o para calcular F1-score
def f1_score_manual(precision, recall):
    return 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

# Fun√ß√£o para calcular acur√°cia
def accuracy_manual(TP, TN, FP, FN):
    return (TP + TN) / (TP + TN + FP + FN)

# Fun√ß√£o para calcular todas as m√©tricas e exibir os resultados
def print_metrics_manual(y_true, y_pred, classifier_name):
    TP, TN, FP, FN = confusion_matrix_manual(y_true, y_pred)
    precision = precision_manual(TP, FP)
    recall = recall_manual(TP, FN)
    f1 = f1_score_manual(precision, recall)
    accuracy = accuracy_manual(TP, TN, FP, FN)

    print(f"--- {classifier_name} ---")
    print("Matriz de Confus√£o:")
    print(f"TP: {TP}, TN: {TN}, FP: {FP}, FN: {FN}")
    print(f"Precis√£o: {precision:.4f}")
    print(f"Revoca√ß√£o: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Acur√°cia: {accuracy:.4f}")
    print("\n")

```

```python
# M√©tricas para Setosa vs Versicolor (Perceptron)
print_metrics_manual(y_setosa_versicolor, predictions_sv_2d, "Perceptron (Setosa vs Versicolor)")

# M√©tricas para Setosa vs Versicolor (KNN)
print_metrics_manual(y_setosa_versicolor, predictions_knn_sv, "KNN (Setosa vs Versicolor)")

# M√©tricas para Setosa vs Virginica (Perceptron com Regra Delta)
print_metrics_manual(y_setosa_virginica, predictions_sv2_delta, "Perceptron com Regra Delta (Setosa vs Virginica)")

# M√©tricas para Setosa vs Virginica (KNN)
print_metrics_manual(y_setosa_virginica, predictions_knn_sv2, "KNN (Setosa vs Virginica)")
```
