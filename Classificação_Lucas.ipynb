```python
import numpy as np
from sklearn.datasets import load_iris

# Função para calcular a distância euclidiana
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# Classe do classificador KNN
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        # Computa as distâncias
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]

        # Obtém os k vizinhos mais próximos
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # Obtém o rótulo de classe mais comum
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        return most_common

# Carregar a base de dados Iris
print(load_iris())
iris = load_iris()
X = iris.data
y = iris.target
class_names = iris.target_names  # Nomes das classes
```

```python
class Perceptron:
    # O método __init__ inicializa o Perceptron com uma taxa de aprendizado e um número de iterações.
    def __init__(self, learning_rate=0.01, n_iter=1000):
        # Armazena a taxa de aprendizado (learning_rate) fornecida.
        self.learning_rate = learning_rate
        # Armazena o número de iterações (n_iter) que o algoritmo executará.
        self.n_iter = n_iter

    # O método fit treina o modelo Perceptron nos dados de entrada X e nos rótulos correspondentes y.
    def fit(self, X, y):
        # Armazena o número de amostras (n_samples) e o número de características (n_features) dos dados de entrada X.
        self.n_samples, self.n_features = X.shape
        # Inicializa os pesos (weights) como um array de zeros, com o mesmo número de elementos que as características.
        self.weights = np.zeros(self.n_features)
        # Inicializa o bias (termo de polarização) como zero.
        self.bias = 0
        # Inicializa uma lista vazia para armazenar o número de erros em cada iteração.
        self.errors_ = []

        # Loop principal do algoritmo de aprendizado, que será executado pelo número de iterações especificado.
        for _ in range(self.n_iter):
            # Inicializa o contador de erros para a iteração atual.
            errors = 0
            # Itera sobre cada amostra x_i e seu respectivo rótulo y[idx].
            for idx, x_i in enumerate(X):
                # Calcula a saída linear, que é o produto escalar dos pesos e a amostra, somado ao bias.
                linear_output = np.dot(x_i, self.weights) + self.bias
                # Aplica a função de ativação (step function) à saída linear para obter a previsão (predicted).
                predicted = self._step_function(linear_output)
                # Calcula o valor de atualização, que depende da diferença entre o rótulo real e a previsão.
                update = self.learning_rate * (y[idx] - predicted)
                # Atualiza os pesos multiplicando o valor de atualização pela amostra.
                self.weights += update * x_i
                # Atualiza o bias somando o valor de atualização.
                self.bias += update
                # Incrementa o contador de erros se houver uma atualização (update não for zero).
                errors += int(update != 0.0)
            # Adiciona o número de erros da iteração atual à lista de erros.
            self.errors_.append(errors)

    # A função step_function aplica a função de ativação degrau (step function) à entrada x.
    def _step_function(self, x):
        # Retorna 1 se x for maior ou igual a 0, caso contrário, retorna 0.
        return np.where(x >= 0, 1, 0)

    # O método predict faz previsões para um conjunto de dados X, usando o modelo treinado.
    def predict(self, X):
        # Calcula a saída linear para cada amostra em X.
        linear_output = np.dot(X, self.weights) + self.bias
        # Aplica a função de ativação e retorna as previsões.
        return self._step_function(linear_output)

# Cria uma instância do Perceptron com uma taxa de aprendizado de 0.01 e 1000 iterações.
perceptron = Perceptron(learning_rate=0.01, n_iter=1000)

# Treina o Perceptron nos dados de entrada X_binary e nos rótulos correspondentes y_binary.
perceptron.fit(X_binary, y_binary)

# Faz previsões usando os dados de entrada X_binary.
predictions = perceptron.predict(X_binary)

# Calcula a acurácia das previsões comparando-as com os rótulos reais y_binary.
accuracy = np.mean(predictions == y_binary)

# Imprime a acurácia.
print("Acurácia:", accuracy)
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Função para plotar a superfície de decisão
def plot_decision_regions(X, y, classifier, title, resolution=0.02):
    # Configuração do marcador e mapa de cores
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # Superfície de decisão
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Plot das amostras
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=cl, edgecolor='black')
    plt.title(title)
    plt.xlabel('Característica 1')
    plt.ylabel('Característica 2')
    plt.legend(loc='upper left')
    plt.show()

# Implementação do Perceptron
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.learning_rate = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):
        self.n_samples, self.n_features = X.shape
        self.weights = np.zeros(self.n_features)
        self.bias = 0
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                predicted = self._step_function(linear_output)
                update = self.learning_rate * (y[idx] - predicted)
                self.weights += update * x_i
                self.bias += update
                errors += int(update != 0.0)
            self.errors_.append(errors)

    def _step_function(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self._step_function(linear_output)

# Implementação do Perceptron com Regra Delta
class PerceptronDelta:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.learning_rate = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y):
        self.n_samples, self.n_features = X.shape
        self.weights = np.zeros(self.n_features)
        self.bias = 0
        self.errors_ = []

        for _ in range(self.n_iter):
            total_error = 0
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                predicted = self._step_function(linear_output)
                error = y[idx] - predicted
                self.weights += self.learning_rate * error * x_i
                self.bias += self.learning_rate * error
                total_error += int(error != 0)
            self.errors_.append(total_error)

    def _step_function(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self._step_function(linear_output)

# Implementação do KNN
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        return most_common

# Função para calcular a distância euclidiana
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# Carregar a base de dados Iris
iris = load_iris()
X = iris.data
y = iris.target

# Normalizar os dados
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Casos separáveis: Iris Setosa vs Versicolor (classes 0 e 1)
binary_indices_sv = y < 2
X_setosa_versicolor = X[binary_indices_sv]
y_setosa_versicolor = y[binary_indices_sv]

# Treinar e avaliar o Perceptron para Setosa vs Versicolor
perceptron_2d = Perceptron(learning_rate=0.01, n_iter=1000)
perceptron_2d.fit(X_setosa_versicolor[:, :2], y_setosa_versicolor)
predictions_sv_2d = perceptron_2d.predict(X_setosa_versicolor[:, :2])
accuracy_sv_2d = np.mean(predictions_sv_2d == y_setosa_versicolor)
print("Acurácia Perceptron (Setosa vs Versicolor) com 2 características:", accuracy_sv_2d)

# Treinar e avaliar o KNN para Setosa vs Versicolor
knn_sv = KNN(k=3)
knn_sv.fit(X_setosa_versicolor[:, :2], y_setosa_versicolor)
predictions_knn_sv = knn_sv.predict(X_setosa_versicolor[:, :2])
accuracy_knn_sv = np.mean(predictions_knn_sv == y_setosa_versicolor)
print("Acurácia KNN (Setosa vs Versicolor):", accuracy_knn_sv)

# Plotar a superfície de decisão para o Perceptron
plot_decision_regions(X_setosa_versicolor[:, :2], y_setosa_versicolor, perceptron_2d, "Superfície de Decisão Perceptron (Setosa vs Versicolor)")

# Casos não separáveis: Iris Setosa vs Virginica (classes 0 e 2)
binary_indices_sv2 = (y == 0) | (y == 2)
X_setosa_virginica = X[binary_indices_sv2]
y_setosa_virginica = y[binary_indices_sv2]

# Treinar e avaliar o Perceptron com Regra Delta para Setosa vs Virginica
perceptron_delta = PerceptronDelta(learning_rate=0.01, n_iter=1000)
perceptron_delta.fit(X_setosa_virginica[:, :2], y_setosa_virginica)
predictions_sv2_delta = perceptron_delta.predict(X_setosa_virginica[:, :2])
accuracy_sv2_delta = np.mean(predictions_sv2_delta == y_setosa_virginica)
print("Acurácia Perceptron com Regra Delta (Setosa vs Virginica):", accuracy_sv2_delta)

# Treinar e avaliar o KNN para Setosa vs Virginica
knn_sv2 = KNN(k=3)
knn_sv2.fit(X_setosa_virginica[:, :2], y_setosa_virginica)
predictions_knn_sv2 = knn_sv2.predict(X_setosa_virginica[:, :2])
accuracy_knn_sv2 = np.mean(predictions_knn_sv2 == y_setosa_virginica)
print("Acurácia KNN (Setosa vs Virginica):", accuracy_knn_sv2)
```

```python
# Função para calcular a matriz de confusão
def confusion_matrix_manual(y_true, y_pred):
    TP = TN = FP = FN = 0
    for true, pred in zip(y_true, y_pred):
        if true == 1 and pred == 1:
            TP += 1
        elif true == 0 and pred == 0:
            TN += 1
        elif true == 0 and pred == 1:
            FP += 1
        elif true == 1 and pred == 0:
            FN += 1
    return TP, TN, FP, FN

# Função para calcular precisão
def precision_manual(TP, FP):
    return TP / (TP + FP) if (TP + FP) != 0 else 0

# Função para calcular revocação (sensibilidade)
def recall_manual(TP, FN):
    return TP / (TP + FN) if (TP + FN) != 0 else 0

# Função para calcular F1-score
def f1_score_manual(precision, recall):
    return 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

# Função para calcular acurácia
def accuracy_manual(TP, TN, FP, FN):
    return (TP + TN) / (TP + TN + FP + FN)

# Função para calcular todas as métricas e exibir os resultados
def print_metrics_manual(y_true, y_pred, classifier_name):
    TP, TN, FP, FN = confusion_matrix_manual(y_true, y_pred)
    precision = precision_manual(TP, FP)
    recall = recall_manual(TP, FN)
    f1 = f1_score_manual(precision, recall)
    accuracy = accuracy_manual(TP, TN, FP, FN)

    print(f"--- {classifier_name} ---")
    print("Matriz de Confusão:")
    print(f"TP: {TP}, TN: {TN}, FP: {FP}, FN: {FN}")
    print(f"Precisão: {precision:.4f}")
    print(f"Revocação: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Acurácia: {accuracy:.4f}")
    print("\n")

```

```python
# Métricas para Setosa vs Versicolor (Perceptron)
print_metrics_manual(y_setosa_versicolor, predictions_sv_2d, "Perceptron (Setosa vs Versicolor)")

# Métricas para Setosa vs Versicolor (KNN)
print_metrics_manual(y_setosa_versicolor, predictions_knn_sv, "KNN (Setosa vs Versicolor)")

# Métricas para Setosa vs Virginica (Perceptron com Regra Delta)
print_metrics_manual(y_setosa_virginica, predictions_sv2_delta, "Perceptron com Regra Delta (Setosa vs Virginica)")

# Métricas para Setosa vs Virginica (KNN)
print_metrics_manual(y_setosa_virginica, predictions_knn_sv2, "KNN (Setosa vs Virginica)")
```
