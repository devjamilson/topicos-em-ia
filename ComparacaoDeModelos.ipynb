### **Matriz de Confusão - Setosa e Versicolor**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Definindo as médias das classes para duas características principais
mean_class1 = vetor_media_setosa[:2]  # Selecionando as duas primeiras características
mean_class2 = vetor_media_versicolor[:2]  # Selecionando as duas primeiras características

# Função para calcular a distância euclidiana
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

# Função para classificar um ponto baseado na distância mínima
def classify_point(point, mean_class1, mean_class2):
    distance_to_class1 = euclidean_distance(point, mean_class1)
    distance_to_class2 = euclidean_distance(point, mean_class2)

    if distance_to_class1 < distance_to_class2:
        return 0  # Classe 1
    else:
        return 1  # Classe 2

# Converter os dados para arrays NumPy e selecionar duas características
if isinstance(data_setosa, pd.DataFrame):
    class1_data = data_setosa.iloc[:, :2].values  # Selecionando duas primeiras características
else:
    class1_data = data_setosa[:, :2]

if isinstance(data_versicolor, pd.DataFrame):
    class2_data = data_versicolor.iloc[:, :2].values  # Selecionando duas primeiras características
else:
    class2_data = data_versicolor[:, :2]

# Dados de teste (substitua pelos seus dados reais se tiver)
test_class1_data = class1_data  # Utilizando os dados de classe 1 como teste
test_class2_data = class2_data  # Utilizando os dados de classe 2 como teste

# Verdadeiras classes
true_labels = np.concatenate([np.zeros(len(test_class1_data)), np.ones(len(test_class2_data))])

# Dados combinados para teste
test_data = np.vstack([test_class1_data, test_class2_data])

# Classificar os dados de teste
predicted_labels = np.array([classify_point(point, mean_class1, mean_class2) for point in test_data])

# Construir a matriz de confusão
cm = confusion_matrix(true_labels, predicted_labels, labels=[0, 1])

# Exibir a matriz de confusão
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Classe 1', 'Classe 2'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusão')
plt.show()
```

# **Perceptron com a Regra Delta**

### **Matriz de Confusão para Setosa e Virginica**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Supondo que 'data' seja um DataFrame já carregado com os dados de íris

# Substituir as vírgulas por pontos nas 4 primeiras colunas
data.iloc[:, :4] = data.iloc[:, :4].replace(',', '.', regex=True)

# Converter as colunas de string para float
data.iloc[:, :4] = data.iloc[:, :4].astype(float)

# Filtrar apenas as classes 'setosa' e 'virginica'
data = data[data.iloc[:, 4].isin(['setosa', 'virginica'])]

# Verificar a distribuição das classes no DataFrame
print("Distribuição das classes no DataFrame original:")
print(data.iloc[:, 4].value_counts())

# Separar as features (X) e a classe (y)
X = data.iloc[:, :4].values  # Pegar todas as linhas e as 4 primeiras colunas
y = data.iloc[:, 4].values  # Pegar todas as classes associadas (5ª coluna)

# Adicionar uma coluna de uns para incluir o bias
ones_column = np.ones((X.shape[0], 1))
X = np.hstack((X, ones_column))

# Separando os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Codificar rótulos de saída para valores numéricos
y_train = np.where(y_train == 'setosa', 1, -1)
y_test = np.where(y_test == 'setosa', 1, -1)

# Verificar a distribuição de classes nos conjuntos de treino e teste
print("Distribuição de classes no conjunto de treino:", np.bincount(y_train + 1))  # Ajustar para índices -1 e 1
print("Distribuição de classes no conjunto de teste:", np.bincount(y_test + 1))

class Perceptron:
    def __init__(self, vectors_size, learn_rate=0.04):
        self.weights = np.zeros(vectors_size, dtype=np.float64)
        self.learn_rate = learn_rate

    def fit(self, X, y, epochs):
        quantidade_amostras = len(X)
        for epoch in range(epochs):
            for i in range(quantidade_amostras):
                dotProduct = self.calculate_dot_product(X[i])
                y_pred = 1 if dotProduct > 0 else -1
                # Atualizar pesos usando a regra delta
                if y[i] != y_pred:  # Atualiza apenas se houve erro de classificação
                    self.weights += self.learn_rate * (y[i] - y_pred) * X[i].astype(np.float64)

    def predict(self, X):
        y_pred = []
        for i in range(len(X)):
            dotProduct = self.calculate_dot_product(X[i])
            y_pred.append(1 if dotProduct > 0 else -1)
        return np.array(y_pred)

    def calculate_dot_product(self, x):
        return np.dot(x, self.weights)

# Instanciar o perceptron e treinar o modelo
perceptron = Perceptron(X_train.shape[1])
perceptron.fit(X_train, y_train, epochs=1000)

# Fazer previsões com o modelo treinado
y_pred = perceptron.predict(X_test)

# Calcular a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)

print("Classes verdadeiras: ", y_test)
print("Classes preditas: ", y_pred)
print("Acurácia:", accuracy)

# Criar e exibir a matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=[-1, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['virginica', 'setosa'])
disp.plot()

# Exibir o gráfico
plt.title('Matriz de Confusão para o Perceptron com Regra Delta')
plt.show()


```

# **Perceptron**

### **Matriz de Confusão para Setosa e Versicolor**

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/arquivo/Iris data - Iris.csv')
data

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Supondo que 'data' seja um DataFrame já carregado com os dados de íris

# Substituir as vírgulas por pontos nas 4 primeiras colunas
data.iloc[:, :4] = data.iloc[:, :4].replace(',', '.', regex=True)

# Converter as colunas de string para float
data.iloc[:, :4] = data.iloc[:, :4].astype(float)

# Filtrar o DataFrame para manter apenas as classes 'setosa' e 'versicolor'
data = data[data.iloc[:, 4].isin(['setosa', 'versicolor'])]

# Separar as features (X) e a classe (y)
X = data.iloc[:, :4].values  # Pegar todas as linhas e as 4 primeiras colunas
y = data.iloc[:, 4].values  # Pegar todas as classes associadas (5ª coluna)

# Mapear as classes para valores numéricos
class_mapping = {'setosa': 1, 'versicolor': -1}
y = np.array([class_mapping[label] for label in y], dtype=np.float64)

# Adicionar uma coluna de uns para incluir o bias
ones_column = np.ones((X.shape[0], 1), dtype=np.float64)
X = np.hstack((X, ones_column))

# Separar os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

class Perceptron:
    def __init__(self, vectors_size):
        self.weights = np.zeros(vectors_size, dtype=np.float64)
        self.learn_rate = 0.01

    def fit(self, X, y, epochs):
        for epoch in range(epochs):
            for i in range(len(X)):
                dot_product = self.calculate_dot_product(X[i])
                # Atualizar os pesos apenas se houver erro
                if y[i] * dot_product <= 0:
                    # Verifique e corrija os tipos de dados para evitar erros
                    X_i = X[i].astype(np.float64)
                    y_i = np.float64(y[i])
                    self.weights += self.learn_rate * (y_i * X_i)

    def predict(self, X):
        return np.array([self.predict_single(x) for x in X])

    def predict_single(self, x):
        return 1 if self.calculate_dot_product(x) > 0 else -1

    def calculate_dot_product(self, x):
        return np.dot(x, self.weights)

# Instanciar o perceptron e treinar o modelo
perceptron = Perceptron(X_train.shape[1])
perceptron.fit(X_train, y_train, epochs=1000)

# Fazer previsões com o modelo treinado
y_pred = perceptron.predict(X_test)

# Calcular a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)

print("Classes verdadeiras: ", y_test)
print("Classes preditas: ", y_pred)
print("Acurácia:", accuracy)

# Criar e exibir a matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=[1, -1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['setosa', 'versicolor'])
disp.plot(cmap=plt.cm.Blues)

# Exibir o gráfico
plt.title('Matriz de Confusão para o Perceptron')
plt.show()

```

## **Coeficiente Kappa - Indica a previsão ajustada pelo acaso**

### **Kappa CDM**

 Calculando o Acerto Aleatório

```python
import numpy as np

# Matriz de confusão
matriz = np.array([[50, 0],
                   [2, 48]])

print("Matriz de Confusão:")
print(matriz)
```

```python
#Calculando o Produto Escalar - Somatória do produto das margens

import numpy as np

# Listas
ac_ = np.array([52, 48])
a_c = np.array([50, 50])

# Calcular o produto escalar
produto = np.dot(ac_, a_c)

print("Somatória do produto dos elementos:", produto)
```

```python
#Caculando o Acerto Aleatório
Aa = produto/(100^2)
print("Acerto Aleatório:", Aa)
```

Aplicando na fórmula do Kappa

```python
K = (0.98 - Aa)/(1 - Aa)
print("O valor do Kappa corresponde a:", K)
```
